```python
import pandas as pd
import plotly.express as px

wine_data = pd.read_csv('wine_qt.csv') # leer los datos
fig = px.histogram(wine_data, x="quality") # crear un histograma
fig.show()
```

```python
fig = px.scatter(wine_data, x="alcohol", y="quality") # crear un gráfico de dispersión
fig.show()
```
